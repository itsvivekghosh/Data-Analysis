```python
df=pd.read_csv("train.csv", usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
def one_hot_top_10(df, variable, top_10_labels):
    
    for label in top_10_labels:
        df[variable+"_"+label]=np.where(df[variable]==label, 1, 0)
        
for col in ['X1', 'X2', 'X3', 'X4', 'X5', 'X6']:
    one_hot_top_10(df, col, top_10)
    df=df.drop(col, axis=1)
```
